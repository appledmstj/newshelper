```python
import re
import csv
```

```python
terminates_front = []

with open('C:\\Users\\Eunseo\\Documents\\GitHub\\newshelper\\front_terminate.csv') as file:
    for line in file.readlines():
        terminates_front.append(line.split(','))
print(terminates_front)
```

```python
terminates_rear = []

with open('C:\\Users\\Eunseo\\Documents\\GitHub\\newshelper\\terminates.csv') as file2:
    for line in file2.readlines():
        terminates_rear.append(line.split(','))
print(terminates_rear)
```

```python
def cleanText(readData):
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', str(readData))
    return text
```

```python
for i in range(len(terminates_front)):
    terminates_front[i] = cleanText(terminates_front[i])
for j in range(len(terminates_rear)):
    terminates_rear[i] = cleanText(terminates_rear[i])
```

```python
new_front = []
for v in terminates_front:
    if v not in new_front:
        new_front.append(v)
print(new_front)
```

```python
new_rear = []
for v in terminates_rear:
    if v not in new_rear:
        new_rear.append(v)
print(new_rear)
```

```python
print(len(new_front[0]), len(new_rear[0]))
```

```python
front = []
for v in new_front:
    front.append(v.split(' '))
rear = []
for v in new_rear:
    rear.append(v.split(' '))
print(len(front[0]))
```

```python
final_front = []
for v in front[0]:
    if v not in final_front:
        final_front.append(v)
print(final_front)
```

```python
final_rear = []
for v in rear[0]:
    if v not in final_rear:
        final_rear.append(v)
print(final_rear)
```

```python
f = open("C:\\Users\\Eunseo\\Documents\\GitHub\\newshelper\\clean_front.csv", 'w', newline='')

writer = csv.writer(f)
writer.writerow(final_front)
f.close()
```

```python
f = open("C:\\Users\\Eunseo\\Documents\\GitHub\\newshelper\\clean_rear.csv", 'w', newline='')

writer = csv.writer(f)
writer.writerow(final_rear)
f.close()
```
